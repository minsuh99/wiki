```python
# SQL INSERT INTO & Insert Data Only in Specific Columns
with sqlite3.connect("./w3schools.db") as conn:
    cur = conn.cursor()
    # ID 빼고 삽입할 내용 지정 -> ID는 자동으로 증가
    cur.execute("INSERT INTO Customers (CustomerName, ContactName, Address, City, PostalCode, Country) " \
    "VALUES ('Cardinal', 'Tom B. Erichsen', 'Skagen 21', 'Stavanger', '4006', 'Norway');")

    # 왜 예제 코드를 적용하면 id가 자동으로 증가하는 기능을 못 쓰는지 모르겠음
    # cur.execute("INSERT INTO Customers " \
    # "VALUES ('Cardinal', 'Tom B. Erichsen', 'Skagen 21', 'Stavanger', '4006', 'Norway');")
    conn.commit() # 데이터 베이스를 변형하기 때문에 .commit()을 통한 변경 내용 적용
    cur.execute("SELECT * FROM Customers WHERE CustomerName = ?", ('Cardinal',))
    rows = cur.fetchall()

    columns = [desc[0] for desc in cur.description]
    print(columns)
    
    for row in rows:
        print(row)
```

```python
# SQL UPDATE
with sqlite3.connect("./w3schools.db") as conn:
    cur = conn.cursor()
    cur.execute("UPDATE Customers "\
                "SET ContactName='Juan' "\
                "WHERE Country='Mexico';") # WHERE 구문 없으면 전체 데이터가 변경됨

    conn.commit()

    cur.execute("SELECT * FROM Customers WHERE CustomerID < 10;")
    rows = cur.fetchall()

    columns = [desc[0] for desc in cur.description]
    print(columns)
    
    for row in rows:
        print(row)
```

```python
# SQL DELETE
# DELETE (table name): 테이블 내용 전체 제거
# DROP (table name): 테이블 전체 구조까지 제거
with sqlite3.connect("./w3schools.db") as conn:
    cur = conn.cursor()
    cur.execute("DELETE FROM Customers WHERE CustomerName='Alfreds Futterkiste';") # 1번 레코드 삭제
    conn.commit()

    cur.execute("SELECT * FROM Customers WHERE CustomerID < 5;")
    rows = cur.fetchall()

    columns = [desc[0] for desc in cur.description]
    print(columns)
    
    for row in rows:
        print(row)
```

```python
# SQL Aggregate Functions
# SQL MIN()
with sqlite3.connect("./w3schools.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT MIN(Price) " \
    "FROM Products;")
    rows = cur.fetchall()

    columns = [desc[0] for desc in cur.description]
    print(columns) # 기본적으로 사용한 agg함수명으로 열 이름 지정

    for row in rows:
        print(row)
```

```python
# Set Column Name (Alias)
with sqlite3.connect("./w3schools.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT MIN(Price) AS SmallestPrice " \
    "FROM Products;")
    rows = cur.fetchall()

    columns = [desc[0] for desc in cur.description]
    print(columns)  # 열 이름이 AS 뒤의 열 이름으로 지정

    for row in rows:
        print(row)
```

```python
# SQL SELECT INTO
# INTO 뒤의 테이블명으로 기존 테이블 내용을 활용해 새 테이블 생성
# SQLite에서는 지원하지 않아 CREATE 사용
with sqlite3.connect("./w3schools.db") as conn:
    cur = conn.cursor()
    # cur.execute("SELECT * INTO CustomersBackup2026 " \
    # "FROM Customers;")
    cur.execute("CREATE TABLE CustomersBackup2026 AS " \
    "SELECT * FROM Customers;")
    cur.execute("SELECT * FROM CustomersBackup2026;")
    conn.commit()
    
    rows = cur.fetchall()

    columns = [desc[0] for desc in cur.description]
    print(columns)

    for row in rows:
        print(row)
```
